```python
# Import libraries and data to df.

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:\Users\Nicolas\Downloads\movies.csv')
```

```python
# Get a glimpse of the data to verify it's loaded correctly.

df.head()
```

```python
# Check for any missing data.

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# For the sake of pacing and simplicity, we are going to drop the missing data columns

df = df.dropna()
```

```python
# We check again.

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Now we verify each column's data type.

df.dtypes
```

```python
# Change numeric columns data type to a more appropriate format.

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# Create a corrected year column.

df['year_corrected'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# Sort by gross revenue.

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Output a list of all the companies without any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Finding correlations in the Data
# Expect budget to have a high correlation with Gross Revenue
# We also believe that company has a somewhat high correlation
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Revenue')
plt.xlabel('Gross Revenue')
plt.ylabel('Budget')

plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Let's start looking at correlation
```

```python
# Note that there are different type of correlation methods, pearson (default), kendall and spearman.

df.corr(method='pearson', numeric_only=True)

# High correlation between budget and gross, as expected
```

```python
correlation_matrix = df.corr(method='pearson', numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Now we want to make a df where all values are numeric, so we transform data like company, country, director into numeric values

df_numerized = df

for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson', numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr = high_corr[(high_corr) < 1]

high_corr
```
